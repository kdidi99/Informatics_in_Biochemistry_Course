```python
import numpy as np
import matplotlib.pyplot as plt
```

#### We start with simple plotting, let's create some data:

```python
x  = np.linspace(0,7,120)
y1 = x**0.5
y2 = x
y3 = x**2
```

#### We can plot this easily...

```python
plt.figure(figsize=(14,7), dpi=320) #figsize is wide and length in inches, dpi is resolution in dots per inch
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
```

#### but we can do it nicer, with the
### Object oriented style
Like everything in python, matplotlib works with objects, in the following, we will see, that we always have an object for the frame figure and an object for every subplot.

```python
class students(object):
    def __init__(self, number, names, knowledge):
        self.number = number
        self.names = names
        self.knowledge = knowledge
    def python_seminar (self, gain):
        self.knowledge += gain

group = students(3, ('Alfred', 'Gertrude', 'Bert'), 0)
print(group.knowledge)
group.python_seminar(100)
print(group.knowledge)
print(type(group))
print(dir(group))
```

```python
fig = plt.figure(figsize=(14,3), dpi=320)
print(type(fig))
ax = fig.add_subplot()
print(type(ax))
plt.show()
```

```python
fig, axes = plt.subplots(figsize=(14, 3), dpi=320)
print(type(fig))
print(type(axes))
plt.show()
```

```python
print("The class students has the methods \n", dir(group), "\n")
print("The class figure has the methods \n", dir(fig), "\n")
print("The class axes has the methods \n", dir(axes))
```
