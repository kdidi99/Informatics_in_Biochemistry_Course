```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x  = np.linspace(0,7,120)
y1 = x**0.5
y2 = x
y3 = x**2
```

Let's take another look on the methods of the axes class.

```python
fig = plt.figure(figsize=(14,5), dpi=160)
ax1 = fig.add_subplot(111)
dir(ax1)
```

So we see that the plot method is only one of many of the methods in matplotlib class axes. We are going to see what some of the other methods are going to do. All these other methods are used to make different plot styles. The big advantage now is, that we can mix all styles in one plot because all the methods belong to the same class.

```python
fig = plt.figure(figsize=(14,5), dpi=160)
ax1 = fig.add_subplot(121, xlim=(0,2), ylim=(0,3))
ax1.bar(x,y1,color='lightseagreen')
ax1.scatter(x,y2,c='green')
ax1.plot(x,y3)
ax2 = fig.add_subplot(1,2,2, xlim=(0,2), ylim=(0,3))
ax2.stem(x,y1)
ax2.step(x,y2,color= 'lightseagreen')
ax2.fill_between(x,y1,y3, color='green')
plt.show()
```

#### We have some methods to visualize 3 dimensional data in 1 dimensional plots

```python
#Let's create some data
x = np.linspace(0,4*np.pi,20)
y = np.linspace(0,4*np.pi,200)
Y,X = np.meshgrid(x,y)
print(X)
print(Y)
Z = np.sin(X)*np.cos(Y)+np.sin(Y)

#And continue with plotting
fig = plt.figure(figsize=(14,5), dpi=160)
ax1 = fig.add_subplot(221)
ax1.pcolormesh(X,Y,Z)
ax1.plot(y,y,c='red')
ax2 = fig.add_subplot(2,2,2)
ax2.imshow(Z)
ax3 = fig.add_subplot(2,2,3)
ax3.contour(X,Y,Z)
ax4 = fig.add_subplot(2,2,4)
ax4.contourf(X,Y,Z)
```

#### We also have methods to plot statistical data

```python
#Let's create some random data
import random
print(random.random())

random_x = []
random_y = []
for i in range(3000):
    random_x.append(random.gauss(0.5,0.3))
    random_y.append(random.random())
    
#print(random_data)
```

```python
fig = plt.figure(figsize=(14,7), dpi=160)

ax1 = fig.add_subplot(221)
ax1.hist(random_x, 200)
ax2 = fig.add_subplot(222)
ax2.errorbar(np.linspace(0,10,10), random_x[0:10], yerr= random_y[0:10])
ax3 = fig.add_subplot(223)
#ax3.hist2d(random_x, random_y,(100,100))
ax3.hexbin(random_x, random_y, gridsize= 20, cmap=plt.get_cmap('Greens'))
ax4 = fig.add_subplot(224)
ax4.pie([6,9,12])
plt.show()
```
