Okey, now we are able to create figures and axes, that will not do the job. So let's finally start with all the nice and beautiful graphic things. We will see our first real plot now and we will clean it up perfectly. Therefore, we start specifying our line properties. 

```python
import numpy as np
import matplotlib.pyplot as plt
import inspect
```

```python
x  = np.linspace(0,7,120)
y1 = x**0.5
y2 = x
y3 = x**2
```

```python
fig, axes = plt.subplots(figsize=(14, 7), dpi=320)
print(axes.viewLim)
print(inspect.signature(axes.plot))
axes.plot(x,np.sin(y1), color='lightseagreen', linewidth= 5, linestyle = 'dotted',label="$\mathrm{sin}(\sqrt{x})$")
#we can use keywords as arguments
print(axes.viewLim)
#we can also use shortforms for linewidth, linestyle and color...
axes.plot(x,-np.sin(y2), 'x-', c='red', lw=1, markersize=10, label="$-\mathrm{sin}(x)$")
#or we can use hexdec color codes
print(axes.viewLim)
axes.plot(x,np.cos(y3), c='#947294', lw=2, ls='--', label="$-\mathrm{cos}(x^2)$")
print(axes.viewLim)
axes.legend(loc=0)
plt.show()
```

We can label everything in matplotlib and we can put labels exactly were we want to.

```python
fig, axes = plt.subplots(figsize=(14,4), dpi=320)
axes.set_xlim(0,5) #We can set the x and y limits of the figure specifically as we want to.
axes.set_ylim(-2,3.435)
axes.set_xlabel("$x$-values", fontsize=15)
axes.set_ylabel("$y$-values", fontsize=15)
axes.plot(x[1:120],1/y1[1:120], label="$\\frac{1}{\sqrt{x}}$")
print(axes.viewLim)
axes.legend(loc='lower left', fontsize=20)
axes.set_title("An example plotting $f(x) = \\frac{1}{\sqrt{x}}$", fontsize=20)
plt.show()
```

That's not all, but for further labeling, we can use coordinates. These are defined very precise, which you can see in the following example. Therefore, we will create a little more data.

```python
t = np.linspace(-15,15,5000)
r = np.cos(t**2)
```

```python
%matplotlib notebook 
#this line will activate the interactive mode, which will allow you to zoom and read coordinates
plt.plot(t,r)
```

```python
%matplotlib inline 
#this line will deactivate interactive mode
coords = np.array([(-2.5,1)])
fig, ax = plt.subplots(figsize=(14,5), dpi=320)
ax.plot(t,r)
ax.text(0,0,"Write something here", rotation='vertical', ha='center', va='center')
#also an angle is possible for rotation, ha or horizontal alignment, va for vertical alignment, we can also insert fontsize keyword as everywere
ax.annotate("This is a peak", xy=coords[0], xytext=coords[0]+(0,0.6), rotation='horizontal', va='center', ha='center',arrowprops=dict(
                    width=0.31,headwidth=5,color= 'black'), fontsize=20)
```

After we know now how to label everything, we can take care of fitting our axes. We start by setting the range and then thinking of the ticks.

```python
fig, ax = plt.subplots(figsize=(14,5), dpi=320)
ax.plot(t,r)
#ax.set_xlim(-5,5) this would do the job, but there is a nicer way to do it when we want to look at both axes
ax.set(xlim=(-5,5), ylim=(0,1))
ax.set(xticks=np.arange(-5,6,1), yticks=[0,1])
ax.tick_params(axis='both', which='major', labelsize=15) #For one axis only use x and y as char.
plt.show()
```

There are two more little parameters which take care of the background:

```python
fig, ax = plt.subplots(figsize=(14,5), dpi=320)
ax.plot(t,r)
ax.set(xlim=(-5,5), ylim=(0,1))
ax.grid(visible='True', which='major', color='green') #same parameters as tick_params
ax.set_facecolor('#847833')
```
