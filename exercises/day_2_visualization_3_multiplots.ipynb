```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x  = np.linspace(0,7,120)
y1 = x**0.5
y2 = x
y3 = x**2
```

```python
fig, axes = plt.subplots(ncols=2, figsize=(14,5), dpi=160)
print(type(axes))
print(type(axes[0]))
axes[0].plot(x,y1)
axes[1].plot(x,y3)
print((fig.get_axes()))
```

#### Another method

```python
fig = plt.figure(figsize=(14,5), dpi=160)
ax1 = fig.add_subplot(121) #number are (height_width_number) in fractions (2 means half width/heigth)
ax2 = fig.add_subplot(122)
ax2.label_outer()
ax1.plot(y1)
ax2.plot(y2)
```

#### We can also stack plots horizontally:

```python
fig, axes = plt.subplots(ncols=3,nrows=2,figsize=(14,7), dpi=160, sharex = True, sharey=True)
axes[0,0].plot(y1)
axes[0,1].plot(y2)
axes[0,2].plot(y3)
axes[1,0].plot(-y1)
axes[1,1].plot(-y2)
axes[1,2].plot(-y3)
```

#### We can also use completly assymetric styles:

```python
from matplotlib import gridspec

fig = plt.figure(figsize=(14,10), dpi=80)
gs = gridspec.GridSpec(4, 4)
print(type(gs))
ax0 = plt.subplot(gs[0,0:3])
ax0.plot(x, y1)
print(type(ax0))

ax1 = plt.subplot(gs[0,3])
ax1.plot(y2, x)
ax2 = plt.subplot(gs[1:4,0])
ax2.plot(y1, x)

ax3 = plt.subplot(gs[1:3,1:3])
ax3.plot(x, y2)
ax4 = plt.subplot(gs[1:4,3])
ax4.plot(y3, x)
ax5 = plt.subplot(gs[3,1])
ax5.plot(y3, x)

ax6 = plt.subplot(gs[3,2])
ax6.plot(x, y1)
```

#### Another thing is to use a shared axes

```python
fig = plt.figure(figsize=(14,5), dpi = 160)
ax1 = fig.add_subplot(111)
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
ax1.tick_params(axis='y', labelcolor='red')
ax2.tick_params(axis='y', labelcolor='blue')
ax1.plot(x,y1,c='red')
ax2.plot(x,y2,c='blue')
```

#### We can also adjust plots to the size we want:

```python
from mpl_toolkits.axes_grid.inset_locator import (inset_axes, InsetPosition,
                                                  mark_inset)
#Import an additional library!

fig = plt.figure(figsize=(14,5), dpi=160)
ax1 = fig.add_subplot(111)
ax2 = plt.axes([0,0,1,1])
ip = InsetPosition(ax1, [0.4,0.2,0.5,0.5])
ax2.set_axes_locator(ip)
print(type(ax1))
print(type(ax2))  #Attetion here, these are objects of different classes
```
