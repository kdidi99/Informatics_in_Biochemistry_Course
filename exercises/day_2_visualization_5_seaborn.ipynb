```python
import numpy as np
total  = np.loadtxt("protein.csv", delimiter=",")
hydrop = np.loadtxt("hydrop.csv", delimiter=",")

hydrophobis = []
for i in range(len(hydrop[0])):
    hydrophobis.append([hydrop[0,i], hydrop[1,i]])

# first import the pandas library
import pandas as pd


# create a list of lists
coords = []
for i in range(len(total[0])):
    coords.append([total[0,i], total[1,i], [total[0,i], total[1,i]] in hydrophobis])
    #print([total[0,i], total[1,i]])

# creating the Pandas DataFrame
df = pd.DataFrame(coords, columns = ['x', 'y', 'Hydrophobic'])

# to display the DataFrame.
df

```
