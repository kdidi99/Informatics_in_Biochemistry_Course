#### After executing the following block, you will have the data of a sodium spectrum with the arrays called wavelength and intensity. Try to plot the spectrum for the wavelengths between 350 and 550 nm and clean it up.
Cleaning it up means make it ready to put it in a paper.<br>
Don't forget to import all the libraries you need.

```python
import numpy as np
def comma_to_float(valstr):
    return float(valstr.decode("utf-8").replace(',','.'))
wavelength, intensity = np.loadtxt('sodium.txt', skiprows=17,converters= {0:comma_to_float, 1:comma_to_float},comments='>', unpack=True)
```

```python
fig, ax = plt.subplots(num=None, figsize=(14, 8.25), dpi=320, facecolor='w', edgecolor='k')

ax.plot(wavelength, intensity/1000, color = 'lightseagreen')
ax.axis([350, 550, 0, 55])
ax.set_xlabel('wavelength [nm]', fontsize= 15)
ax.set_ylabel('intensity [1000]', fontsize = 15)
ax.tick_params(axis='both', which='major', labelsize=15)
ax.grid(visible='True', which='major')
ax.set(yticks=[10,20,30,40,50])
plt.show()
```

Now copy your solution into the cell below and try to add labels to the peaks. An example is shown at the end of the page. In the follwing cell, the indices of the peaks are given.

```python
peaks = np.array([1094, 1113, 1147, 1161, 1179, 1205, 1264, 1284, 1340, 1381, 1471, 1494, 1578])
```

```python
fig, ax = plt.subplots(num=None, figsize=(14, 7), dpi=80, facecolor='w', edgecolor='k')

for i in range(len(peaks)):
    plt.annotate(wavelength[peaks[i]], xy=(wavelength[peaks[i]], intensity[peaks[i]]/1000+0.1),
                 xytext = (wavelength[peaks[i]]-1, intensity[peaks[i]]/1000+4), 
                 rotation=90, color = 'lightseagreen',
                 arrowprops=dict(
                    width=0.31,headwidth=5,color= 'lightseagreen'))

ax.plot(wavelength, intensity/1000, color = 'black')
ax.axis([350, 550, 1, 60])
ax.set_xlabel('wavelength [nm]', fontsize= 15)
ax.set_ylabel('intensity [1000]', fontsize = 15)
ax.tick_params(axis='both', which='major', labelsize=15)
ax.grid(visible='True', which='major')
ax.set(yticks=[10,20,30,40,50])
plt.show()
```
