```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
total  = np.loadtxt("protein.csv", delimiter=",")
hydrop = np.loadtxt("hydrop.csv", delimiter=",")
```

In the cell above we the coordinates of the amino acids of a protein. All we have is the coordinates of all amino acids in the array "total" and coordinates of the hydrophobic amino acids in the array "hydrop". The x-coordinates are in the first entry and the y-coordinates are in the second entry.<br>
Can you decide if the protein is soluble in water or would rather lay in a membrane region?<br>
Therefore, try to make a statistic plot to decide where you can find the amino acids. You can use a subplot variant with shared axes to get a better overview or you can add marking lines in your histogram to make a decission of the solubility. You can also try to do both.

```python
#method1
fig, ax = plt.subplots(figsize=(14,8), dpi=160)
ax.scatter(total[0], total[1], 50, alpha=0.2, label = "total")
ax.scatter(hydrop[0], hydrop[1], 50, alpha = 0.2, color='red', label = "hydrophobic")
ax.set(xlim=(-6.5,6.5), ylim=(-5.5,5.5))
ax.legend()
```

```python
#method2
figure, axes = plt.subplots(ncols=2, figsize=(14,7), dpi = 160, sharex = True, sharey = True)
axes[0].hexbin(total[0], total[1], gridsize= 50, cmap=plt.get_cmap('Greens'))
axes[1].hexbin(hydrop[0], hydrop[1], gridsize= 50, cmap=plt.get_cmap('Greys'))

axes[0].set_xlabel("$x$-values", fontsize=15)
axes[1].set_xlabel("$x$-values", fontsize=15)
axes[0].set_ylabel("$y$-values", fontsize=15)
axes[0].set_xlim(-5,5)
axes[0].set_ylim(-7.5,7.5)
axes[1].set_xlim(-5,5)
axes[1].set_ylim(-7.5,7.5)
axes[0].set_title("Amino acid distribution", fontsize=20)
axes[1].set_title("Distribution of hydrophobic amino acids", fontsize=20)
```
