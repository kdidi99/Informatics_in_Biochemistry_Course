```python
import numpy as np
import matplotlib.pyplot as plt
```

Create a figure containing 4 plots.
1. $f_1(x) = \mathrm{sin}(x)$
2. $f_2(x) = \mathrm{cos}(x)$
3. $f_3(x) = \mathrm{sinh}(x)$
4. $f_4(x) = \mathrm{cosh}(x)$<br>

First try to plot all 4 functions in 4 different plots with shared x-axes. In the second cell try to plot sinus and sinus hyperbolicus in the same plot with shared axes as well as cosinus and cosinus hyperbolicus in a second plot in the same figure below the sinus plot.<br>
<br>
If you still have time, try to play around with colours and labels.

```python
x    = np.linspace(-np.pi,np.pi,120)
sin  = np.sin(x)
cos  = np.cos(x)
sinh = np.sinh(x)
cosh = np.cosh(x)
```

```python
fig, axes = plt.subplots(ncols=2,nrows=2, figsize=(14,5), dpi=160, sharex = True)
axes[0,0].plot(x,sin)
axes[0,1].plot(x,cos)
axes[1,0].plot(x,sinh)
axes[1,1].plot(x,cosh)
```

```python
fig = plt.figure(figsize=(14,5), dpi = 160)
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
ax3 = ax1.twinx()
ax4 = ax2.twinx()
ax1.tick_params(axis='y', labelcolor='red')
ax3.tick_params(axis='y', labelcolor='blue')
ax2.tick_params(axis='y', labelcolor='orange')
ax4.tick_params(axis='y', labelcolor='green')
ax1.plot(x,sin,c='red')
ax2.plot(x,cos,c='orange')
ax3.plot(x,sinh,c='blue')
ax4.plot(x,cosh,c='green')
```
